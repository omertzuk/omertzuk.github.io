## Jupyter Notebooks in the Garden

This is an example showing how Jupyter notebooks can be part of the digital garden. Notebooks are perfect for:

- Exploratory data analysis
- Interactive visualizations
- Documenting computational experiments
- Teaching concepts with live code

Let's create a simple visualization as a placeholder.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a simple plot
plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='sin(x)', linewidth=2)
plt.plot(x, y2, label='cos(x)', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Trigonometric Functions')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Why Notebooks?

Notebooks allow for:

- **Reproducible research**: Code, results, and narrative in one place
- **Iterative development**: Easy to modify and re-run
- **Rich output**: Plots, tables, interactive widgets
- **Documentation**: Markdown cells for context and explanation

This is just a simple example. Future notebooks might explore:

- EEG time-series analysis
- Complex systems simulations
- Network analysis and visualization
- Machine learning experiments
- Knowledge graph queries
